```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
from sqlalchemy import create_engine
```

```python
# Read the combined data CSV file
combined_df = pd.read_csv('data/combined_data.csv')
combined_df.head()
```

```python
# Create SQLAlchemy engine
engine = create_engine('sqlite:///claimed_time.db')
# Connect to the database
conn = engine.connect()
```

```python
# Save the DataFrame to a table in the database
combined_df.to_sql('claimed_table', con=conn, if_exists='replace', index=False)
```

```python
# Query all records in the database
claimed_df = pd.read_sql("Select * from claimed_table", conn)
claimed_df
```
