```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
claimed_df = pd.read_excel("data/AppServices_AllClaimingWorkSegments.xlsx")

claimed_df.head()
```

```python
# Drop unnecessary columns
claimed_df = pd.read_excel("data/AppServices_AllClaimingWorkSegments.xlsx", usecols=['EPM_ProductionControlItemID', 'EPM_InstanceNumber',
                                                                                     'TimeInSeconds_ThisWorkSegment', 'NumberWithDash',
                                                                                    'EPM_MainMark', 'EPM_AdjustedStationName',
                                                                                    'Assembly_MainPartLengthFt', 'Assembly_WeightEachLbs',
                                                                                    'Assembly_SurfaceAreaEachSqFt', 'Assembly_MainPieceProductionCode',
                                                                                    'Assembly_TotalQuantityInJob', 'Assembly_MainPartShape',
                                                                                    'Assembly_MainPartDimension', 'Assembly_MainPartFinishDescr',
                                                                                    'Assembly_NumSmallParts', 'Assembly_EstTotalHours_ThisLaborGroup'])

claimed_df.head()
```

```python
# Reorder columns
claimed_df = claimed_df[['NumberWithDash', 'EPM_ProductionControlItemID',  'EPM_MainMark', 'EPM_InstanceNumber', 'Assembly_MainPieceProductionCode',
           'Assembly_MainPartLengthFt', 'Assembly_WeightEachLbs', 'Assembly_SurfaceAreaEachSqFt','Assembly_MainPartShape', 'Assembly_MainPartDimension',
           'Assembly_MainPartFinishDescr', 'Assembly_TotalQuantityInJob', 'Assembly_NumSmallParts', 'EPM_AdjustedStationName', 'TimeInSeconds_ThisWorkSegment',
           'Assembly_EstTotalHours_ThisLaborGroup']]

claimed_df.head()
```

```python
# change Fab/Layout - Check in the EPM_AdjustedStationName column to Fab/Layout, change weld check to weld
modified_claimed_df = claimed_df
modified_claimed_df['EPM_AdjustedStationName'] = modified_claimed_df['EPM_AdjustedStationName'].replace({'Fab/Layout - Check': 'Fab/Layout',
                                                                                                         'Weld Check': 'Weld',
                                                                                                         'Fab After Galv - Check' : 'Fab After Galv'})
modified_claimed_df.head()
```

```python
# combine rows if the NumberWithDash, EPM_ProductionControlItemID, EPM_MainMark, EPM_InstanceNumber, and EPM_AdjustedStationName columns are the same. 
# Add the numbers in the TimeInSeconds_ThisWorkSegment, and	Assembly_EstTotalHours_ThisLaborGroup columns when combining

combined_df = modified_claimed_df.groupby(['NumberWithDash', 'EPM_ProductionControlItemID', 'EPM_MainMark', 'EPM_InstanceNumber',
                                           'Assembly_MainPieceProductionCode', 'Assembly_MainPartLengthFt', 'Assembly_WeightEachLbs',
                                           'Assembly_SurfaceAreaEachSqFt', 'Assembly_MainPartShape', 'Assembly_MainPartDimension',
                                           'Assembly_MainPartFinishDescr', 'Assembly_TotalQuantityInJob', 'Assembly_NumSmallParts',
                                           'EPM_AdjustedStationName']).agg({
    'TimeInSeconds_ThisWorkSegment': 'sum',
    'Assembly_EstTotalHours_ThisLaborGroup': 'sum'
}).reset_index()

combined_df
```

```python
#Assembly_NumSmallParts includes the main part so subtract 1 from this column for accurate count
combined_df['Assembly_NumSmallParts'] -= 1
combined_df
```

```python
# Save the combined DataFrame to a CSV file in the 'data' folder
import os
data_folder = 'data'
if not os.path.exists(data_folder):
    os.makedirs(data_folder)
    
#combined_df.to_csv('combined_data.csv', index=False)
combined_df.to_csv(os.path.join(data_folder, 'combined_data.csv'), index=False)
```
